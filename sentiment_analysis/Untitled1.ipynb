```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("taylor_swift_spotify.csv")

# Convert release_date to datetime format
data['release_date'] = pd.to_datetime(data['release_date'])

# Extract year from release_date
data['year'] = data['release_date'].dt.year

# Group by year and calculate the mean valence for each year
yearly_avg_valence = data.groupby('year')['valence'].mean()

# Plotting
plt.figure(figsize=(12, 6))
yearly_avg_valence.plot(marker='o', color='green', linestyle='-')
plt.title("Average Valence of Taylor Swift's Songs Over the Past 10 Years")
plt.xlabel("Year")
plt.ylabel("Average Valence")
plt.grid(True)
plt.xticks(yearly_avg_valence.index)  # Set xticks to match years
plt.show()
```
