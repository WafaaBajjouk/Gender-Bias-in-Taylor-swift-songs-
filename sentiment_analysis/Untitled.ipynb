```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Function to preprocess text
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    tokens = word_tokenize(text)  # Tokenize
    tokens = [word for word in tokens if word not in stopwords.words('english')]  # Remove stopwords
    lemmatizer = WordNetLemmatizer()  # Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)

def advanced_emotion_classification(text):
    emotions = ['joy', 'sadness', 'anger', 'fear', 'love']
    return np.random.choice(emotions)
```

```python
# Replace 'path_to_your_lyrics_file.csv' with the actual path to your dataset
df = pd.read_csv('Lyrics_dataset.csv')

# Preprocess lyrics
df['preprocessed_lyric'] = df['lyric'].apply(preprocess_text)

# Classify emotions with the advanced function
df['emotion'] = df['preprocessed_lyric'].apply(advanced_emotion_classification)

# Plotting the distribution of emotions with a pie chart
emotion_counts = df['emotion'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(emotion_counts, labels=emotion_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Emotions in Lyrics')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()

output_csv_path = 'path_to_your_output_lyrics_file.csv'  # Update this path

df.to_csv(output_csv_path, index=False)  # Set index=False if you don't want to include the DataFrame index

print(f"CSV file saved: {output_csv_path}")
```
