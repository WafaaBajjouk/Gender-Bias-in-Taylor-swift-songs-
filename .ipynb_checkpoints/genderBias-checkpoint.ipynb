# Gender Bias in Taylor Swift Songs

I will Analyze gender bias in the lyrics column and group the results by each different album

### Methodology
the gender bias analysis in song lyrics is done using NLP tools .

```python

male_terms = [
    'he', 'him', 'his', 'himself',
    'boy', 'boyfriend', 'husband', 'father', 'son', 'uncle', 'nephew', 'brother',
    'man', 'men', 'male', 'gentleman', 'gentlemen', 'sir', 'mr', 'mister', 'king', 'prince', 'lord'
]
female_terms = [
    'she', 'her', 'hers', 'herself',
    'girl', 'girlfriend', 'wife', 'mother', 'daughter', 'aunt', 'niece', 'sister',
    'woman', 'women', 'female', 'lady', 'ladies', 'miss', 'ms', 'mrs', 'queen', 'princess', 'madam'
]

import pandas as pd
import re

df = pd.read_csv('cleaned_lyrics_dataset.csv')

neutral_pronouns = ['they', 'them', 'theirs', 'themselves']  # Added 'themselves' for completeness

# Update function name and use regex for more accurate counting
def count_gender_terms(text, terms):
    """Count the occurrences of specified terms in the text using regex for whole word match, case insensitive."""
    count = 0
    for term in terms:
        # Pattern to match whole word, case insensitive
        pattern = r'\b' + term + r'\b'
        count += len(re.findall(pattern, text, flags=re.IGNORECASE))
    return count

# Apply the updated function
df['male_terms'] = df['lyric'].apply(lambda x: count_gender_terms(x, male_terms) if isinstance(x, str) else 0)
df['female_terms'] = df['lyric'].apply(lambda x: count_gender_terms(x, female_terms) if isinstance(x, str) else 0)
df['neutral_pronouns'] = df['lyric'].apply(lambda x: count_gender_terms(x, neutral_pronouns) if isinstance(x, str) else 0)

# Group by album_name and sum
album_gender_bias = df.groupby('album_name').agg({
    'male_terms': 'sum',
    'female_terms': 'sum',
    'neutral_pronouns': 'sum'
}).reset_index()

album_gender_bias['total_terms'] = album_gender_bias['male_terms'] + album_gender_bias['female_terms'] + album_gender_bias['neutral_pronouns']

album_gender_bias

```

# Visualisation of the result 

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8))
ind = range(len(album_gender_bias))
width = 0.35       

p1 = ax.bar(ind, album_gender_bias['male_terms'], width, color='blue', label='Male terms')
p2 = ax.bar(ind, album_gender_bias['female_terms'], width, bottom=album_gender_bias['male_terms'], color='pink', label='Female termss')
p3 = ax.bar(ind, album_gender_bias['neutral_pronouns'], width, bottom=album_gender_bias['male_pronouns'] + album_gender_bias['female_pronouns'], color='green', label='Neutral Pronouns')

ax.set_xlabel('Albums')
ax.set_ylabel('Counts of Pronouns')
ax.set_title('Counts of Gender-Specific Pronouns by Album')
ax.set_xticks(ind)
ax.set_xticklabels(album_gender_bias['album_name'], rotation='vertical')
ax.legend()

plt.subplots_adjust(bottom=0.3)
plt.show()
```
