```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('taylor_swift_spotify.csv') 

relevant_columns = ['album', 'release_date', 'acousticness', 'instrumentalness']
df = df[relevant_columns]

df['release_date'] = pd.to_datetime(df['release_date'])

df_grouped = df.groupby('album').mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(df_grouped['release_date'], df_grouped['acousticness'], marker='o', label='Acousticness')
plt.plot(df_grouped['release_date'], df_grouped['instrumentalness'], marker='o', label='Instrumentalness')

plt.xlabel('Release Date')
plt.ylabel('Value')
plt.title("Taylor Swift's Vocal Tone Evolution")
plt.legend()

plt.xticks(rotation=45)

```
